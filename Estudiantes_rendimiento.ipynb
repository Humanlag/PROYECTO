```python
import pandas as pd
import numpy as np

# Número de estudiantes
n = 150

# Generación de datos aleatorios
data = {
    'ID': range(1, n+1),
    'Edad': np.random.randint(18, 30, size=n),
    'Horas_estudio': np.random.randint(5, 20, size=n),
    'Asistencia': np.random.randint(60, 100, size=n),
    'Participacion_actividades': np.random.choice([0, 1], size=n),
    'Promedio_calificaciones': np.random.uniform(5.0, 10.0, size=n).round(2),
    'Motivacion': np.random.choice([1, 2, 3], size=n),
    'Acceso_recursos': np.random.choice([0, 1], size=n),
    'Tiempo_plataforma': np.random.uniform(1, 10, size=n).round(1),
    'Rendimiento': np.random.choice([1, 2, 3], size=n)  # 1 = Alto, 2 = Medio, 3 = Bajo
}

# Crear DataFrame
df = pd.DataFrame(data)

# Guardar en formato CSV en el entorno de Colab
csv_file_path = '/content/estudiantes_rendimiento.csv'
df.to_csv(csv_file_path, index=False)

# Descargar el archivo CSV a tu máquina local
from google.colab import files
files.download(csv_file_path)
```

```python
# Importar las bibliotecas necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Subir el archivo desde tu computadora
from google.colab import files
uploaded = files.upload()

# Cargar el archivo CSV
df = pd.read_csv('estudiantes_rendimiento.csv')  # Asegúrate de usar el nombre correcto del archivo subido

# 1. Mostrar las primeras filas del DataFrame
print("Primeras filas del dataset:")
print(df.head())

# 2. Descripción general de las columnas
print("\nDescripción general de las columnas:")
print(df.info())

# 3. Descripción estadística (para columnas numéricas)
print("\nDescripción estadística de las columnas numéricas:")
print(df.describe())

# 4. Analizar la existencia de valores nulos
print("\nCantidad de valores nulos por columna:")
print(df.isnull().sum())

# 5. Distribución de las variables
# Histograma de la Edad
plt.figure(figsize=(8, 6))
df['Edad'].hist(bins=10, edgecolor='black')
plt.title('Distribución de la Edad de los Estudiantes')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()

# Histograma de las Horas de Estudio
plt.figure(figsize=(8, 6))
df['Horas_estudio'].hist(bins=10, edgecolor='black', color='orange')
plt.title('Distribución de las Horas de Estudio de los Estudiantes')
plt.xlabel('Horas de Estudio por Semana')
plt.ylabel('Frecuencia')
plt.show()

# Gráfico de barras: Rendimiento Académico
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Rendimiento')
plt.title('Distribución del Rendimiento Académico')
plt.xlabel('Rendimiento Académico (1 = Alto, 2 = Medio, 3 = Bajo)')
plt.ylabel('Cantidad de Estudiantes')
plt.show()

# 6. Análisis de correlación entre variables numéricas
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlación entre Variables Numéricas')
plt.show()

# 7. Explicación de cada columna
column_explanations = {
    "ID": "Identificador único para cada estudiante.",
    "Edad": "Edad del estudiante, expresada en años.",
    "Horas_estudio": "Promedio de horas que el estudiante dedica al estudio por semana.",
    "Asistencia": "Porcentaje de asistencia del estudiante a clases.",
    "Participacion_actividades": "Indica si el estudiante participa en actividades extracurriculares (1 = Sí, 0 = No).",
    "Promedio_calificaciones": "Promedio de calificaciones del último semestre, de 0 a 10.",
    "Motivacion": "Nivel de motivación del estudiante (1 = Baja, 2 = Media, 3 = Alta).",
    "Acceso_recursos": "Indica si el estudiante tiene acceso a recursos como internet o biblioteca (1 = Sí, 0 = No).",
    "Tiempo_plataforma": "Promedio de horas que el estudiante pasa en plataformas educativas en línea cada semana.",
    "Rendimiento": "Clasificación del rendimiento académico del estudiante (1 = Alto, 2 = Medio, 3 = Bajo)."
}

# Imprimir las explicaciones de cada columna
for col, explanation in column_explanations.items():
    print(f"\n{col}: {explanation}")

```
